```python
capA = 4
capB = 3
goal = 2

visited = set()
stack = [(0, 0, ["Start: JugA=0, JugB=0"])]

while stack:
    a, b, steps = stack.pop()

    if (a, b) in visited:
        continue
    visited.add((a, b))

    if a == goal or b == goal:
        print("Goal reached!\nSteps:\n")
        for s in steps:
            print(s)
        break

    if a < capA:
        stack.append((capA, b, steps + [f"Fill Jug A. Now: ({capA},{b})"]))
    if b < capB:
        stack.append((a, capB, steps + [f"Fill Jug B. Now: ({a},{capB})"]))
   
    if a > 0:
        stack.append((0, b, steps + [f"Empty Jug A. Now: (0,{b})"]))
    if b > 0:
        stack.append((a, 0, steps + [f"Empty Jug B. Now: ({a},0)"]))
  
    if a > 0 and b < capB:
        pour = min(a, capB - b)
        newA = a - pour
        newB = b + pour
        stack.append((newA, newB, steps + [f"Pour water from A to B. Now: ({newA},{newB})"]))
    if b > 0 and a < capA:
        pour = min(b, capA - a)
        newA = a + pour
        newB = b - pour
        stack.append((newA, newB, steps + [f"Pour water from B to A. Now: ({newA},{newB})"]))
```
