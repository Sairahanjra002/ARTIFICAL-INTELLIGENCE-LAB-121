```python
n = int(input("Enter the number of queens (N): "))

board = [[" " for _ in range(n)] for _ in range(n)]
placed = []  

def show_board():
    line = "+" + "----+" * n
    for i in range(n):
        print(line)
        for j in range(n):
            print(f"| {board[i][j]} ", end="")
        print("|")
    print(line)

show_board()

def is_safe(row, col):
    for qr, qc in placed:
        if qr == row or qc == col or abs(qr - row) == abs(qc - col):
            return False
    return True

while True:
    if len(placed) == n:
        print("All queens placed successfully! You won!")
        break

    pos = input("Enter position as row,col (or 'q' to quit): ")

    if pos.lower() == "q":
        print("Exiting game.")
        break

    try:
        r, c = map(int, pos.split(","))
    except:
        print("Invalid input. Use row,col format.")
        continue

    if r < 0 or r >= n or c < 0 or c >= n:
        print("Position outside board. Try again.")
        continue

    if not is_safe(r, c):
        print("Cannot place queen here! Conflicts with existing queens.")
        continue

    if board[r][c] == "Q":
        print("Queen already placed here!")
        continue

    board[r][c] = "Q"
    placed.append((r, c))
    print("Queen placed successfully!")

    show_board()
```
